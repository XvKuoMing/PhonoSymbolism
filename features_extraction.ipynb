```python
consonants = {
    'soft_palate': ['к', 'г'],
    'active_articulator': ['п', 'б', 'м'],
    'upper_teeth': ['т', 'д', 'н']
}

vowels = {
    'back': ['а', 'о', 'у'],
    'front': ['и', 'е']
}
```

```python
import pandas as pd

df = pd.read_csv("responses.csv", sep=',')
df['concept'] = df['associations'] + df['image']
df = df.drop(['associations', 'image'], axis=1)
df
```

```python
features_map = {**consonants, **vowels}
features = {column:[] for column in (features_map).keys()}
features_map = {value: key
                for key, values in features_map.items()
                for value in values
                }

def update_features(text: str) -> None:
  for key in features.keys():
    features[key].append(0)
  for char in text:
    if char in features_map.keys():
      features[features_map[char]][-1] += 1
```

```python
from numpy import vectorize

vectorize(update_features)(df['concept'])

df = df.drop(['concept'], axis=1)
df = df.join(pd.DataFrame(features))
df
```

```python
df.to_csv('responses_with_features.csv')
```
